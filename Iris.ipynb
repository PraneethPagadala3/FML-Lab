```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
iris=pd.read_csv('IRIS.csv')
```

```python
iris.shape
```

```python
iris.head()
```

```python
iris.species.unique()
```

```python
iris["species"].value_counts()
```

```python
x=iris["sepal_length"].values
y=iris["sepal_width"].values
```

```python
plt.scatter(x,y)
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.show()
```

```python
sns.jointplot(x="sepal_length",y="sepal_width",data=iris,size=5)
plt.show()
```

```python
sns.stripplot(y="sepal_length",x="species",data=iris,size=6)
plt.show()
```

```python
sns.boxplot(x="sepal_length",y="species",data=iris)
plt.show()
```

```python
sns.violinplot(x="species",y="petal_length",data=iris,size=5)
plt.show()
```

```python
iris.hist(edgecolor='red',bins=25,figsize=(15,7))
plt.show()
```

```python
sns.pairplot(data=iris,hue="species",size=4)
plt.show()
```

```python
sns.heatmap(iris.corr(),cmap="YlGnBu",annot=True,fmt="f")
plt.show()
```
